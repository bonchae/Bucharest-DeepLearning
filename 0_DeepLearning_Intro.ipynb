# Deep Learning

## When to Use Machine Learning vs Deep Learning

| Criteria | Machine Learning (ML) | Deep Learning (DL) |
|:---|:---|:---|
| **Data Size** | Small to Medium | Large |
| **Data Type** | Structured | Unstructured (image, audio, text) |
| **Compute Needs** | Low to Medium | High (GPU/TPU needed) |
| **Training Time** | Faster | Longer |
| **Feature Engineering** | Manual | Automatic |
| **Interpretability** | Easier | Harder |
| **Example** | Predict house prices | Translate English to French |


## Comparison between Supervised and Unsupervised Deep Learning

| Aspect | Supervised Deep Learning | Unsupervised Deep Learning |
|:---|:---|:---|
| **Need Labels?** | ✅ Yes | ❌ No |
| **Goal** | Map inputs to known outputs | Find patterns or structure |
| **Data Type** | Labeled datasets (input-output pairs) | Unlabeled datasets (inputs only) |
| **Common Tasks** | Classification, Regression | Clustering, Dimensionality Reduction, Data Generation |
| **Example Models** | CNNs, RNNs, Transformers (fine-tuned) | Autoencoders, GANs, Self-supervised models |
| **Evaluation** | Compare predictions to true labels (accuracy, MSE, etc.) | Harder: Evaluate using clustering metrics, reconstruction error, or qualitative judgment |
| **Examples in Practice** | Spam detection, Image recognition, Language translation | Customer segmentation, Anomaly detection, Data compression |
